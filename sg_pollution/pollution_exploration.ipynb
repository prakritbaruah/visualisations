```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.dates import date2num
import datetime
```

```python
def time_in_range(start, end, x):
    """Return true if x is in the range [start, end]"""
    if start <= end:
        return start <= x <= end
    else:
        return start <= x or x <= end
```

```python
rainfall_df = pd.read_csv("data/rainfall-monthly-highest-daily-total.csv")
```

```python
rainfall_df["month"] = pd.to_datetime(rainfall_df["month"])
rainfall_df["maximum_rainfall_in_a_day"] = rainfall_df["maximum_rainfall_in_a_day"].astype(float)
```

```python
# plot every year
years_of_interest = [i for i in range(2010, 2019)]

for year in years_of_interest:
    start_date = datetime.date(year,1,1)
    end_date = datetime.date(year,12,31)
    
    year_rain_df = rainfall_df.loc[
        rainfall_df["month"].apply(lambda x: time_in_range(start_date, end_date, x))
    ]

    labels = [i for i in range(1, 13)]
    rainfall = np.array(year_rain_df["maximum_rainfall_in_a_day"])
    plt.plot(labels, rainfall, linestyle='-')
    
plt.show()
```
